```python
%%sql
CREATE TABLE IF NOT EXISTS stock_performance (
    ticker TEXT,
    price_before_game REAL,
    price_week_after_game REAL,
    trading_volume_millions INTEGER
) STRICT;
```

```python
%%sql
INSERT INTO stock_performance (ticker, price_before_game, price_week_after_game, trading_volume_millions)
VALUES
  ('PEP', 168.50, 172.10, 45),
  ('GOOGL', 142.80, 145.20, 110),
  ('AMZN', 170.20, 178.45, 95),
  ('VZ', 40.10, 39.85, 60),
  ('TSLA', 190.00, 185.20, 150),
  ('TMUS', 197.68, 199.34, 100),
  ('STLA', 22.40, 23.10, 15);
```

```python
%%sql
CREATE Ad_Buys (
    ticker TEXT PRIMARY KEY,
    company_name TEXT,
    ad_cost_millions REAL,
    airtime_seconds INTEGER
) STRICT;
```

```python
%%sql
INSERT INTO ad_buys (ticker, company_name, ad_cost_millions, airtime_seconds)
VALUES
  ('PEP', 'PepsiCo', 7.0, 30),
  ('GOOGL', 'Alphabet (Google)', 14.0, 60),
  ('AMZN', 'Amazon', 14.0, 60),
  ('VZ', 'Verizon', 7.0, 30),
  ('TSLA', 'Tesla', 0.0, 0), -- Tesla doesn't buy ads
  ('TMUS', 'Tmobile', 10.0, 30),
  ('STLA', 'Stellantis (Jeep)', 6.5, 30);
```

```python
%%sql
SELECT *
FROM stock_performance;
```

```python
%%sql
SELECT ticker, price_before_game
FROM stock_performance
ORDER BY ticker ASC;
```

```python
%%sql
SELECT ticker, price_week_after_game
FROM stock_performance
WHERE price_week_after_game > 150;
```

```python
%%sql
SELECT *
FROM ad_buys;
```

```python
%%sql
SELECT ticker, company_name
FROM ad_buys
ORDER BY company_name DESC;
```

```python
%%sql
SELECT ticker, airtime_seconds
FROM ad_buys
WHERE ticker LIKE 'T%'
  AND airtime_seconds = 0;
```
