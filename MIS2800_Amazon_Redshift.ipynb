## YOUR ANSWERS TO PART 2 BELOW

Question 1: For analytics, columnar storage in Amazon Redshift matters more than perfectly normalized tables because analysts pull specific columns across huge datasets, not rows. This storage lets the system scan only what is needed and makes the quaries way faster and more efficient. The performance and speed matter more then transactional systems to help flow smoothly.

Question 2: The leader node handles quary planning, then splits the work across multiple nodes that can process the data all at once. This dramatically reduces how long large reports take to run. Instead of one machine doing all the work, the load is distributed and then helps speed up insights.
